```python
# stackoverflow.com/questions/35488717/confused-about-conv2d-transpose
import tensorflow as tf
```

```python
sess = tf.InteractiveSession()
batch_size = 32
output_shape = [batch_size,7,7,512]
strides = [1,1,1,1]

#const
current = tf.constant(0.1,shape = [batch_size,7,7,1])
w = tf.constant(0.1,shape = [2,2,512,1])

h1 = tf.nn.conv2d_transpose(current,w,output_shape=output_shape,strides=strides,padding='SAME')
print sess.run(h1)
# wrong
```

```python
output = tf.constant(0.1,shape = output_shape)
expected_l = tf.nn.conv2d(output,w,strides = [1,1,1,1],padding='SAME')
print expected_l.get_shape()
```
